## 數值轉換與填補缺失值
* 以平均值 填補缺失Age (轉換數值資料)
* 不參考Cabin (缺漏過多)
* 未參考船票編號. 姓名
* 以眾數 填補Embarked (轉換類別資料)

## 特徵選取
* 性別 (女性生存率較高)
* 年紀 (年紀小生存率高)
* 出發港口 (C港出發生存率高)
* 艙等 (艙等高生存率高)
* 票價 (票價高生存率高)
* 親屬 (直系&旁系分布近似, 有親屬存活率較高)

````python
# 資料分析與前處理

df['Family_size'] = df['Parch'] + df['SibSp']

train_x = df[['Sex', 'Age', 'Embarked', 'Pclass', 'Fare', 'Family_size']]       # 取出訓練資料需要分析的資料欄位
train_y = df['Survived']                       # 取出訓練資料的答案

# train_x.info()
             
from sklearn.impute import SimpleImputer       # 匯入填補缺失值的工具
from sklearn.preprocessing import LabelEncoder # 匯入 Label Encoder

imputer = SimpleImputer(strategy='mean')     # 創造 imputer 並設定填補策略
age = train_x['Age'].to_numpy().reshape(-1, 1)
imputer.fit(age)                               # 根據資料學習需要填補的值
train_x['Age'] = imputer.transform(age)        # 填補缺失值

imputer = SimpleImputer(strategy='most_frequent')
embarked = train_x['Embarked'].to_numpy().reshape(-1, 1)
imputer.fit(embarked)                               
train_x['Embarked'] = imputer.transform(embarked)        

le = LabelEncoder()                            # 創造 Label Encoder
le.fit(train_x['Sex'])                         # 給予每個類別一個數值
train_x['Sex'] = le.transform(train_x['Sex'])  # 轉換所有類別成為數值

le.fit(train_x['Embarked'])                         
train_x['Embarked'] = le.transform(train_x['Embarked'])  
````

## 模型訓練

請在此區塊後使用**一個或多個區塊**進行模型訓練，**需要**包含以下內容：

- **選擇模型**

- 計算**準確度（Accuracy）**

## Decision Tree
* 簡介: 以Information gain(E.g. Gini, Entropy)判斷最具意義的特徵做二元分類
* 本實驗略為Overfitting（訓練與驗證準確度差距 > 10% ）

````python
# 模型訓練
from sklearn.model_selection import GridSearchCV

from sklearn.model_selection import KFold             # 匯入 K 次交叉驗證工具
from sklearn.tree import DecisionTreeClassifier       # 匯入決策樹模型
from sklearn.metrics import accuracy_score            # 匯入準確度計算工具

kf = KFold(n_splits=5,                                # 設定 K 值
           random_state=1012,
           shuffle=True)
kf.get_n_splits(train_x)                              # 給予資料範圍

train_acc_list = []                                   # 儲存每次訓練模型的準確度
valid_acc_list = []                                   # 儲存每次驗證模型的準確度

for train_index, valid_index in kf.split(train_x):    # 每個迴圈都會產生不同部份的資料
    train_x_split = train_x.iloc[train_index]         # 產生訓練資料
    train_y_split = train_y.iloc[train_index]         # 產生訓練資料標籤
    valid_x_split = train_x.iloc[valid_index]         # 產生驗證資料
    valid_y_split = train_y.iloc[valid_index]         # 產生驗證資料標籤
    
    model = DecisionTreeClassifier(random_state=1012, criterion='gini', splitter='random') # 創造決策樹模型
    model.fit(train_x_split, train_y_split)           # 訓練決策樹模型
    
    train_pred_y = model.predict(train_x_split)       # 確認模型是否訓練成功
    train_acc = accuracy_score(train_y_split,         # 計算訓練資料準確度
                               train_pred_y)
    valid_pred_y = model.predict(valid_x_split)       # 驗證模型是否訓練成功
    valid_acc = accuracy_score(valid_y_split,         # 計算驗證資料準確度
                               valid_pred_y)
    
    train_acc_list.append(train_acc)
    valid_acc_list.append(valid_acc)

print((
    'average train accuracy: {}\n' +
    '    min train accuracy: {}\n' +
    '    max train accuracy: {}\n' +
    'average valid accuracy: {}\n' +
    '    min valid accuracy: {}\n' +
    '    max valid accuracy: {}\n').format(
    np.mean(train_acc_list),                          # 輸出平均訓練準確度
    np.min(train_acc_list),                           # 輸出最低訓練準確度
    np.max(train_acc_list),                           # 輸出最高訓練準確度
    np.mean(valid_acc_list),                          # 輸出平均驗證準確度
    np.min(valid_acc_list),                           # 輸出最低驗證準確度
    np.max(valid_acc_list)                            # 輸出最高驗證準確度
))
print('Final accuracy: {}'.format(accuracy_score(train_y, model.predict(train_x))))

parameters = {'random_state': [1012], 'criterion': ['gini', 'entropy'], 'splitter': ['best','random']}
decision_tree_model = DecisionTreeClassifier()
clf = GridSearchCV(decision_tree_model, parameters)
clf.fit(train_x, train_y)
print(clf.best_params_)
````

## Random Forest
* 簡介: 用隨機方式建立一個森林，森林裡面有很多的決策樹，每一棵決策樹之間沒有關聯，最後以決策樹投票結果作為解答。


````python
# 模型訓練

from sklearn.ensemble import RandomForestClassifier       # 匯入隨機森林模型

kf = KFold(n_splits=5,                                # 設定 K 值
           random_state=1012,
           shuffle=True)
kf.get_n_splits(train_x)                              # 給予資料範圍

train_acc_list = []                                   # 儲存每次訓練模型的準確度
valid_acc_list = []                                   # 儲存每次驗證模型的準確度

for train_index, valid_index in kf.split(train_x):    # 每個迴圈都會產生不同部份的資料
    train_x_split = train_x.iloc[train_index]         # 產生訓練資料
    train_y_split = train_y.iloc[train_index]         # 產生訓練資料標籤
    valid_x_split = train_x.iloc[valid_index]         # 產生驗證資料
    valid_y_split = train_y.iloc[valid_index]         # 產生驗證資料標籤
    
    pred_model =  RandomForestClassifier(random_state=1012, criterion='entropy', n_estimators=50, min_samples_split=20, min_samples_leaf=1, max_depth=8)# 創造隨機森林模型
    pred_model.fit(train_x_split, train_y_split)           # 訓練隨機森林模型
    
    train_pred_y = pred_model.predict(train_x_split)       # 確認模型是否訓練成功
    train_acc = accuracy_score(train_y_split,         # 計算訓練資料準確度
                               train_pred_y)
    valid_pred_y = pred_model.predict(valid_x_split)       # 驗證模型是否訓練成功
    valid_acc = accuracy_score(valid_y_split,         # 計算驗證資料準確度
                               valid_pred_y)
    
    train_acc_list.append(train_acc)
    valid_acc_list.append(valid_acc)

print((
    'average train accuracy: {}\n' +
    '    min train accuracy: {}\n' +
    '    max train accuracy: {}\n' +
    'average valid accuracy: {}\n' +
    '    min valid accuracy: {}\n' +
    '    max valid accuracy: {}').format(
    np.mean(train_acc_list),                          # 輸出平均訓練準確度
    np.min(train_acc_list),                           # 輸出最低訓練準確度
    np.max(train_acc_list),                           # 輸出最高訓練準確度
    np.mean(valid_acc_list),                          # 輸出平均驗證準確度
    np.min(valid_acc_list),                           # 輸出最低驗證準確度
    np.max(valid_acc_list)                            # 輸出最高驗證準確度
))
print('Final accuracy: {}'.format(accuracy_score(train_y, pred_model.predict(train_x))))

parameters = {'random_state': [1012],'n_estimators': [10, 20, 50, 100], 'criterion': ['entropy', 'gini'], 'min_samples_split': [20, 40, 80], 'min_samples_leaf': [1, 3, 10], 'max_depth': [4, 6, 8]}
random_forest_model = RandomForestClassifier()
clf = GridSearchCV(random_forest_model, parameters)
clf.fit(train_x, train_y)
print(clf.best_params_)
````

## AdaBoost
* 簡介: 將分類能力比較弱的基礎分類器按照訓練出來的組合係數組合成強分類器
      (前一個基本分類器分錯的樣本會得到加強，加權後的全體樣本再次被用來訓練下一個基本分類器)。


````python
# 模型訓練

from sklearn.ensemble import AdaBoostClassifier       # 匯入模型

kf = KFold(n_splits=5,                                # 設定 K 值
           random_state=1012,
           shuffle=True)
kf.get_n_splits(train_x)                              # 給予資料範圍

train_acc_list = []                                   # 儲存每次訓練模型的準確度
valid_acc_list = []                                   # 儲存每次驗證模型的準確度

for train_index, valid_index in kf.split(train_x):    # 每個迴圈都會產生不同部份的資料
    train_x_split = train_x.iloc[train_index]         # 產生訓練資料
    train_y_split = train_y.iloc[train_index]         # 產生訓練資料標籤
    valid_x_split = train_x.iloc[valid_index]         # 產生驗證資料
    valid_y_split = train_y.iloc[valid_index]         # 產生驗證資料標籤
    
    model =  AdaBoostClassifier(random_state=1012, n_estimators=100, algorithm='SAMME.R')    # 創造模型
    model.fit(train_x_split, train_y_split)           # 訓練模型
    
    train_pred_y = model.predict(train_x_split)       # 確認模型是否訓練成功
    train_acc = accuracy_score(train_y_split,         # 計算訓練資料準確度
                               train_pred_y)
    valid_pred_y = model.predict(valid_x_split)       # 驗證模型是否訓練成功
    valid_acc = accuracy_score(valid_y_split,         # 計算驗證資料準確度
                               valid_pred_y)
    
    train_acc_list.append(train_acc)
    valid_acc_list.append(valid_acc)

print((
    'average train accuracy: {}\n' +
    '    min train accuracy: {}\n' +
    '    max train accuracy: {}\n' +
    'average valid accuracy: {}\n' +
    '    min valid accuracy: {}\n' +
    '    max valid accuracy: {}').format(
    np.mean(train_acc_list),                          # 輸出平均訓練準確度
    np.min(train_acc_list),                           # 輸出最低訓練準確度
    np.max(train_acc_list),                           # 輸出最高訓練準確度
    np.mean(valid_acc_list),                          # 輸出平均驗證準確度
    np.min(valid_acc_list),                           # 輸出最低驗證準確度
    np.max(valid_acc_list)                            # 輸出最高驗證準確度
))
print('Final accuracy: {}'.format(accuracy_score(train_y, model.predict(train_x))))

parameters = {'random_state': [1012], 'n_estimators': [10, 20, 50, 100, 200], 'algorithm': ['SAMME', 'SAMME.R']}
adaboost_model = AdaBoostClassifier()
clf = GridSearchCV(adaboost_model, parameters)
clf.fit(train_x, train_y)
print(clf.best_params_)
````

## SVC
* 簡介: 找出一個超平面(hyperplane)，使兩個不同的集合分開，且希望此平面距離兩集合越大越好，
        透過Kernel function將特徵映射到高維空間可進行非線性分類。
* 本實驗略為underfitting (採用PassengerId 作為唯一特徵)

````python
# 模型訓練

from sklearn.svm import SVC       # 匯入支援向量機模型

train_newx = df[['PassengerId']]       # 取出訓練資料需要分析的資料欄位

kf = KFold(n_splits=5,                                # 設定 K 值
           random_state=1012,
           shuffle=True)
kf.get_n_splits(train_newx)                              # 給予資料範圍

train_acc_list = []                                   # 儲存每次訓練模型的準確度
valid_acc_list = []                                   # 儲存每次驗證模型的準確度

for train_index, valid_index in kf.split(train_newx):    # 每個迴圈都會產生不同部份的資料
    train_x_split = train_newx.iloc[train_index]         # 產生訓練資料
    train_y_split = train_y.iloc[train_index]         # 產生訓練資料標籤
    valid_x_split = train_newx.iloc[valid_index]         # 產生驗證資料
    valid_y_split = train_y.iloc[valid_index]         # 產生驗證資料標籤
    
    model =  SVC(random_state=1012, kernel='rbf', C=1, gamma='scale')                   # 創造支援向量機模型
    model.fit(train_x_split, train_y_split)           # 訓練支援向量機模型
    
    train_pred_y = model.predict(train_x_split)       # 確認模型是否訓練成功
    train_acc = accuracy_score(train_y_split,         # 計算訓練資料準確度
                               train_pred_y)
    valid_pred_y = model.predict(valid_x_split)       # 驗證模型是否訓練成功
    valid_acc = accuracy_score(valid_y_split,         # 計算驗證資料準確度
                               valid_pred_y)
    
    train_acc_list.append(train_acc)
    valid_acc_list.append(valid_acc)

print((
    'average train accuracy: {}\n' +
    '    min train accuracy: {}\n' +
    '    max train accuracy: {}\n' +
    'average valid accuracy: {}\n' +
    '    min valid accuracy: {}\n' +
    '    max valid accuracy: {}').format(
    np.mean(train_acc_list),                          # 輸出平均訓練準確度
    np.min(train_acc_list),                           # 輸出最低訓練準確度
    np.max(train_acc_list),                           # 輸出最高訓練準確度
    np.mean(valid_acc_list),                          # 輸出平均驗證準確度
    np.min(valid_acc_list),                           # 輸出最低驗證準確度
    np.max(valid_acc_list)                            # 輸出最高驗證準確度
))
print('Final accuracy: {}'.format(accuracy_score(train_y, model.predict(train_newx))))

parameters = {'random_state': [1012], 'kernel': ['rbf'], 'C': [1, 10, 100, 1000], 'gamma': ['auto', 'scale']}
svm_model = SVC()
clf = GridSearchCV(svm_model, parameters)
clf.fit(train_x, train_y)
print(clf.best_params_)
````
