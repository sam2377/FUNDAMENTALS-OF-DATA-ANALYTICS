```python
import numpy as np
import pandas as pd

df_train = pd.read_csv('Stock_train.csv')
df_test = pd.read_csv('Stock_test.csv')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style("whitegrid")
plt.plot(df_train['Close Price'], 'black')
```

```python
df_train["NEW_DATE"] = pd.to_datetime(df_train["Date"], format="%d-%b-%Y")
df_train["NEW_DATE"] = pd.to_datetime(df_train["NEW_DATE"]).astype(np.int64)
df_train["Increase"] = df_train["Close Price"] - df_train["Open Price"]
df_train["Amplitude"] = df_train["High Price"] - df_train["Low Price"]
df_train["Movement"] = df_train["Close Price"] - df_train["Open Price"] > 0

df_test["NEW_DATE"] = pd.to_datetime(df_test["Date"], format="%d-%b-%Y")
df_test["NEW_DATE"] = pd.to_datetime(df_test["NEW_DATE"]).astype(np.int64)
df_test["Increase"] = df_test["Close Price"] - df_test["Open Price"]
df_test["Amplitude"] = df_test["High Price"] - df_test["Low Price"]
df_test["Movement"] = df_test["Close Price"] - df_test["Open Price"] > 0

# df_train["Movement"]
```

```python
df_train
```

```python
df_test
```

```python
for i in reversed(range(df_train.shape[0] - 1)):
    if i > 0:
        df_train["Increase"][i] = df_train["Increase"][i-1]
        df_train["Amplitude"][i] = df_train["Amplitude"][i-1]

for i in reversed(range(df_test.shape[0] - 1)):
    if i > 0:
        df_test["Increase"][i] = df_test["Increase"][i-1]
        df_test["Amplitude"][i] = df_test["Amplitude"][i-1]
        
for i in range(df_train.shape[0] - 1):
    df_train["Movement"][i] = df_train["Movement"][i+1]
df_train = df_train.drop([0, df_train.shape[0] - 1])
for i in range(df_test.shape[0] - 1):
    df_test["Movement"][i] = df_test["Movement"][i+1]
df_test = df_test.drop([0, df_test.shape[0] - 1])
```

```python
df_train
```

```python
df_test
```

```python
train_x = df_train[["Open Price", "Close Price", "High Price", "Low Price", "Volume", "Increase", "Amplitude"]]
train_y = df_train["Movement"]

test_x = df_test[["Open Price", "Close Price", "High Price", "Low Price", "Volume",  "Increase", "Amplitude"]]
test_y = df_test["Movement"]

for col in ["Open Price", "Close Price", "High Price", "Low Price", "Volume", "Increase", "Amplitude"]:
    mean = np.average(train_x[col])
    std = np.std(train_x[col])
    train_x[col] = (train_x[col] - mean)/std
    
for col in ["Open Price", "Close Price", "High Price", "Low Price", "Volume", "Increase", "Amplitude"]:
    mean = np.average(test_x[col])
    std = np.std(test_x[col])
    test_x[col] = (test_x[col] - mean)/std
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score

run_gs = True

if run_gs:
    parameter_grid = {
                 "C":np.logspace(-3,3,7),
                 "random_state": [128],
                 }
    logistic = LogisticRegression()
    cross_validation = StratifiedKFold(n_splits=5)

    grid_search = GridSearchCV(logistic,
                               scoring='accuracy',
                               param_grid=parameter_grid,
                               cv=cross_validation,
                               verbose=1
                              )

    grid_search.fit(train_x, train_y)
    model = grid_search
    parameters = grid_search.best_params_

    print('Best score: {}'.format(grid_search.best_score_))
    print('Best parameters: {}'.format(grid_search.best_params_))
else: 
    parameters = {"C":[1.0],
                 "random_state": [128],
                }
logistic = LogisticRegression(**parameters)
kf = KFold(n_splits=5,
           random_state=128,
           shuffle=True)
kf.get_n_splits(train_x)

train_acc_list = []
valid_acc_list = []

for train_index, valid_index in kf.split(train_x):
    train_x_split = train_x.iloc[train_index]
    train_y_split = train_y.iloc[train_index]
    valid_x_split = train_x.iloc[valid_index]    
    valid_y_split = train_y.iloc[valid_index] 
    
    logistic.fit(train_x_split, train_y_split)
    
    train_pred_y = logistic.predict(train_x_split)
    train_acc = accuracy_score(train_y_split,         
                               train_pred_y)
    valid_pred_y = logistic.predict(valid_x_split)
    valid_acc = accuracy_score(valid_y_split,         
                               valid_pred_y)
    
    train_acc_list.append(train_acc)
    valid_acc_list.append(valid_acc)

print((
    'average train accuracy: {}\n' +
    '    min train accuracy: {}\n' +
    '    max train accuracy: {}\n' +
    'average valid accuracy: {}\n' +
    '    min valid accuracy: {}\n' +
    '    max valid accuracy: {}').format(
    np.mean(train_acc_list),                          
    np.min(train_acc_list),                           
    np.max(train_acc_list),                           
    np.mean(valid_acc_list),                          
    np.min(valid_acc_list),                           
    np.max(valid_acc_list)
))
print('Final Train Accuracy: {}'.format(accuracy_score(train_y, logistic.predict(train_x))))
print('Final Test Accuracy: {}'.format(accuracy_score(test_y, logistic.predict(test_x))))
```

```python
from sklearn.svm import SVC

run_gs = True

if run_gs:
    parameter_grid ={
                 'random_state': [128],
                 'kernel': ['rbf','sigmoid'], 
                 'C': [1, 10, 100], 
                 'gamma': ['auto', 'scale']
                 }
    svc = SVC()
    cross_validation = StratifiedKFold(n_splits=5)

    grid_search = GridSearchCV(svc,
                               scoring='accuracy',
                               param_grid=parameter_grid,
                               cv=cross_validation,
                               verbose=1,
                               n_jobs=4
                              )

    grid_search.fit(train_x, train_y)
    model = grid_search
    parameters = grid_search.best_params_

    print('Best score: {}'.format(grid_search.best_score_))
    print('Best parameters: {}'.format(grid_search.best_params_))
else: 
    parameters = {"C":[1.0],
                 "random_state": [128],
                }
svc = SVC(**parameters)
kf = KFold(n_splits=5,
           random_state=128,
           shuffle=True)
kf.get_n_splits(train_x)

train_acc_list = []
valid_acc_list = []

for train_index, valid_index in kf.split(train_x):
    train_x_split = train_x.iloc[train_index]
    train_y_split = train_y.iloc[train_index]
    valid_x_split = train_x.iloc[valid_index]    
    valid_y_split = train_y.iloc[valid_index] 
    
    svc.fit(train_x_split, train_y_split)
    
    train_pred_y = svc.predict(train_x_split)
    train_acc = accuracy_score(train_y_split,         
                               train_pred_y)
    valid_pred_y = svc.predict(valid_x_split)
    valid_acc = accuracy_score(valid_y_split,         
                               valid_pred_y)
    
    train_acc_list.append(train_acc)
    valid_acc_list.append(valid_acc)

print((
    'average train accuracy: {}\n' +
    '    min train accuracy: {}\n' +
    '    max train accuracy: {}\n' +
    'average valid accuracy: {}\n' +
    '    min valid accuracy: {}\n' +
    '    max valid accuracy: {}').format(
    np.mean(train_acc_list),                          
    np.min(train_acc_list),                           
    np.max(train_acc_list),                           
    np.mean(valid_acc_list),                          
    np.min(valid_acc_list),                           
    np.max(valid_acc_list)
))
print('Final Train Accuracy: {}'.format(accuracy_score(train_y, svc.predict(train_x))))
print('Final Test Accuracy: {}'.format(accuracy_score(test_y, svc.predict(test_x))))
```

```python
from keras import models, layers, optimizers, regularizers
from keras.callbacks import EarlyStopping
epochs = 10
batch_size = 32
```

```python
model = models.Sequential()

model.add(layers.Dense(input_dim=7, units=10, activation='relu'))
model.add(layers.Dense(input_dim=10, units=10, activation='relu'))
model.add(layers.Dense(input_dim=10, units=10, activation='relu'))
model.add(layers.Dense(input_dim=10, units=1, activation='relu'))

model.compile(loss='binary_crossentropy', optimizer='rmsprop', metrics=['accuracy'])
es = EarlyStopping(monitor='val_loss', patience=2)

train_history = model.fit(train_x, train_y, epochs=epochs, batch_size=batch_size, validation_split=0.2)

accuracy_train = model.evaluate(train_x, train_y, batch_size=batch_size)[1]
accuracy_test = model.evaluate(test_x, test_y, batch_size=batch_size)[1]
print('Training accuracy: %s' % accuracy_train)
print('Testing accuracy: %s' % accuracy_test)
```

```python
score = model.evaluate(test_x, test_y, verbose=0)
print('test loss:', score[0])
print('test acc:', score[1])

loss = train_history.history['loss']
val_loss = train_history.history['val_loss']
epochs = len(loss)
plt.plot(range(epochs), loss, marker='.', label='loss(training data)')
plt.plot(range(epochs), val_loss, marker='.', label='val_loss(evaluation data)')
plt.legend(loc='best')
plt.grid()
plt.xlabel('epoch')
plt.ylabel('loss')
plt.show()
```
